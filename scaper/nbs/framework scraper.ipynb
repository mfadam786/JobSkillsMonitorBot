```python
import pandas as pd
import numpy as np
import requests
import lxml
from lxml import etree
```

```python
from io import StringIO, BytesIO
```

```python
url = "https://www.thoughtworks.com/radar/a-z"
```

```python
r = requests.get(url)
```

```python
parser = etree.HTMLParser()
tree   = etree.parse(StringIO(r.text), parser)
```

```python
xp= '//*[@id="a-z"]//a/text()'
```

```python
tree.xpath(xp)
```

```python
df = pd.DataFrame(data = tree.xpath(xp), columns=["framework"])
```

```python
df.to_csv("./data/frameworks.csv")
```
