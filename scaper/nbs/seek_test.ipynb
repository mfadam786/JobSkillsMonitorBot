```python
import requests
import lxml.html

def get_seek_pages():

    end_of_pages = False
    START_URL = "https://www.seek.co.nz/jobs-in-information-communication-technology"

    urls = [START_URL, ]


    url = START_URL
    count = 1

    while not end_of_pages:

        res = requests.get(url)

        if res.ok:
            doc = lxml.html.fromstring(res.content)
            next_url = doc.xpath("//a[text()='Next']")
            if next_url:
                url = f"https://www.seek.co.nz{next_url[0].values()[0]}"
                print(url)
                print(count)
                count += 1
                urls.append(url)
            else:
                end_of_pages = True


        else:
            raise ValueError(f"Error parsing url {START_URL}")


get_seek_pages()
```
